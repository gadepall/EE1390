##Data Generation

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
N = 100  #No of samples
x = np.linspace(-10,10,N) #x coordinates
noise_samples = np.random.RandomState(5) #include noise samples
y = 5*np.sin(2*x) + noise_samples.uniform()

plt.plot(x,y,'o')
plt.show()
```

###Split into train and test datasets

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

###Using optimize.curve_fit of scipy for the curve fitting 

```
from scipy import optimize
from IPython.display import display, Math

def test_func(x,amp, omega,phi):
    return amp * np.sin(omega * x + phi)

params, params_covariance = optimize.curve_fit(test_func, x, y, p0=[2, 2, 2])

```

```
plt.figure(figsize=(6, 5))
plt.scatter(x, y, label='Data')
plt.plot(x, test_func(x, params[0], params[1],params[2]),
         label='Fitted function')

plt.legend(loc='best')

plt.show()
```
